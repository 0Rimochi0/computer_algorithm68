```python
Q = "Press Q = Exit"
M = "Press M = Main"
S = "Press S = Start in Program"
state = True
while(state):
    print(Q)
    print(M)
    print(S)

    inp = input("Menu Selected : ")
    if(inp == 'Q'):
        state = False
    elif(inp == 'M'):
        print("Home Menu")
    elif(inp == 'S'):
        try:
            #ให้ผู้ใช้กรอกจำนวนรอบที่ต้องการ
            rounds_str = input("กรุณาระบุจำนวนรอบ (สูงสุด 10 รอบ): ")
            rounds = int(rounds_str)
            
            #ตรวจสอบเงื่อนไข: ถ้ากรอกมาเกิน 10 ให้ปรับเหลือแค่ 10 รอบเท่านั้น
            if rounds > 10:
                print(f"คุณใส่มา {rounds} รอบ เกินลิมิต! ระบบจะปรับให้เหลือ 10 รอบนะครับ")
                rounds = 10
            
            print(f"เริ่มวิ่งอย่างสนุกสนานกันเลย {rounds} รอบ เห้ออ...")
            
            #เริ่มวิ่งตามจำนวน rounds ที่ผ่านการตรวจสอบแล้ว
            for i in range(1, rounds + 1):
                print(f"วิ่งรอบที่ : {i}")
            
            print("เย้!! วิ่งครบแล้ว กลับไปหน้าเมนูกันเลย")
            
        except ValueError:
            print("ผิดพลาด! กรุณาใส่ตัวเลขเท่านั้น อย่าใส่ตัวอักษรนะครับ ไม่งั้นผมจะใส่หน้าคุณ")
    else:
        print("Home Menu")
```
