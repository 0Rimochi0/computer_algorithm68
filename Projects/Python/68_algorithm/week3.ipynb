```python
a = 1
b = 2

print("Sum : ", a + b)
print("Minus : ", a - b)
print("Multiply : ", a * b)
print("Div : ", a / b)
print("? : ", a % b) #หารเอาเศษ ไว้หาเลขคี่
```

```python
a = 3
b = 2

if(a > b):
    print("A มากกกว่า B")
elif(a == b):
    print("A เท่ากันกับ B")
else:
    print("A น้อยกว่า B")
```

```python
name = input("Name is : ")
lname = input("Lname is : ")
score = input("Score is : ")

score = int(score)
if(score >= 80):
    print(name,lname,"A สุดยอดมาก! เรียนไปทำไมเยอะแยะเนี่ย?")
elif(score >= 75):
    print(name,lname,"B+ เก่งมาก ๆ แต่สมองไหล!")
elif(score >= 70):
    print(name,lname,"B ดีงามตามมาตรฐาน ไม่เครียดเกินไป")
elif(score >= 65):
    print(name,lname,"C+ ถึงจะเหนื่อยแต่ก็ยังรอดมาได้ ถือว่ายังไม่ตาย!")
elif(score >= 60):
    print(name,lname,"C เฉียดฉิวสุด ๆ แค่ผ่านก็บุญแล้ว")
elif(score >= 55):
    print(name,lname,"D+ ผ่านแบบเส้นยาแดงผ่าแปด!")
elif(score >= 50):
    print(name,lname,"D บุญมีแต่กรรมบัง!")
else:
    print(name,lname,"F คุณโง่แต่ไว้รอบหน้าเอาใหม่ ถ้ามี?")
```
